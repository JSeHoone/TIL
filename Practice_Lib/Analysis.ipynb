```python
import pandas as pd
import numpy as np
```

## Numpy

```python
# numpy repeat
# Array에 있는 데이터 원소들을 몇번 중복할 것인가 만들어 주는 것
a = np.array([1,2,3,4])
b = a.repeat(2)
c = a.repeat(3)
print(a)
print(b)
print(c)
```

```python
# numpy tile
# Array 자체를 복사해서 뒤에 붙인다.
a = np.array([1,2,3,4])
b = np.tile(a,2)
c = np.tile(a,3)
print(a)
print(b)
print(c)
```

```python
# numpy reshape
# 원소의 총 갯수를 통해 matrix(행렬)을 바꿔준다.
a = np.array([1,2,3,4])
b = np.tile(a,2)
reshape_ = b.reshape((2,4))
print(a.shape)
print(b.shape)
print(reshape_.shape)
```

```python
a = {
    'Name' : ['John', 'Say', 'Hou', 'Gomni'],
    'Age' : [15,20,26,40],
    'Sex' : ["male", 'female', 'male','female']
}

b = {
    'Name' : ['John', 'Somi', 'Hou', 'Chomi'],
    'Salary' : [0,1200,5000,12000],
    'Sex' : ['male', 'female','male','male']
}

df_1 = pd.DataFrame.from_dict(a)
df_2 = pd.DataFrame.from_dict(b)
```

```python
# 데이터 결합 방법 2 ( pd.merge )
display(pd.merge(df_1,df_2, how='inner', on = 'Name')) # 두 Data Frame 에서 Name이라는 column을 inner join하겠다는 의미
display(pd.merge(df_1, df_2, how='left', on = ['Name', 'Sex'])) # 두 Data Frame 에서 Name과 Sex column을 left join하겠다는 의미
```

```python
## crosstab을 이용한 교차표 만들기
display(pd.crosstab(index = df['Sex'], columns= df['Age'], normalize = 'columns'))
# 행은 Data Frame에 있는 성별 column을, 열은 Data Fraem 에 있는 Age column으로 교차표를 생성한다는 의미
# normalize = 'columns'의 Argument는 열별로 백분율을 만들어 준다는 의미 ( 3.00 column을 보면 female이 33퍼, male이 66퍼의 비율을 가진다는 의미)
# index나 columns Argument에 여러 column으로 보고 싶다면 python list형태로 넣어주면 된다.
display(pd.crosstab(index= [df['Sex'], df['Fare']], columns=df['Age'], normalize='columns'))
```

```python
## pivot Table 만들기
display(pd.pivot_table(data = df, index = 'Sex', columns= 'Age', values='Fare', aggfunc='mean'))
# data는 어떤 Data Frame을 사용할 건지 정하는 것이고, index는 테이블에 어떤 columns을 넣을건지, columns 위치에 어떤 column을 넣을 건지
# valuse는 두 columns(특징)들을 통해서 어떤 값을 볼 것인지
# aggfunc는 value에 들어간 값들을 어떤 연산 처리를 할건지 의미한다. (집계값 연산)
```

```python
## Data Frame에 있는 여러가지 데이터 형태
# category 데이터 타입
df = pd.DataFrame({"value": np.random.randint(0, 100, 20)}) # value라는 column에 0~100사이의 랜덤 숫자를 20개 생성한 Data Frame을 만들겠다 ! 
labels = ["{0} - {1}".format(i, i + 9) for i in range(0, 100, 10)] # 0 ~ 100 사이를 step 10을 주어서 리스트를 만들겠다는 의미
df["group"] = pd.cut(df.value, range(0, 105, 10), right=False, labels=labels) 
#                    해당 값이, 해당 범위에 있다면, 범위 오른쪽을 안 포함하고, label값을 넣어주겠다!
df["group_2"] = pd.cut(df.value, range(0, 105, 10), right=True, labels=labels) 
df
```

```python
## Cut 함수에서 Right = False에 대한 추가 설명
print(list(range(0,105,10))) # Range를 뽑아보면 이렇게 보이는데 
# 위에서 bin값에 Range를 넣어준 것이다. 이 의미는 0 ~ 10, 10~20 이렇게 범위를 만든 것인데
# 여기서 오른쪽 값인 0~10인 구간에서 10을 포함 할건지 안할건지에 대해서 물어 보는 것이다. 
df[df.index == 8]
# group은 right포함 안한거고, group_2는 포함한 결과이다.
```
